#Data scaliing


```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

non_numeric = df.select_dtypes(exclude=['number']).columns

for col in non_numeric:
    df[col] = le.fit_transform(df[col].astype(str))

print("Remaining non-numeric columns:")
print(df.select_dtypes(exclude=['number']).columns)
```

```python
# Cell to ensure no NaNs remain before scaling
# If df.info() showed no NaNs, but KMeans complains, this is a robust step.
# We'll fill any remaining NaNs with the mean for numerical columns.
# For any object columns (though all should be encoded by now), it would need mode.

# Re-identifying numerical columns after all encoding
numeric_cols_final = df.select_dtypes(include=['int64', 'float64']).columns

# Filling NaNs in these final numerical columns
df[numeric_cols_final] = df[numeric_cols_final].fillna(df[numeric_cols_final].mean())

print("Final check for NaNs before scaling:")
print(df.isnull().sum().sum())

if df.isnull().sum().sum() == 0:
    print("No NaNs found. Data is ready for scaling.")
else:
    print("WARNING: NaNs still present after final fillna. Please inspect your data.")
```

```python
import numpy as np

# Check for infinite values in df
if np.isinf(df).sum().sum() > 0:
    print("Infinite values found. Replacing with NaN and filling with mean...")
    # Replace infinite values with NaN
    df.replace([np.inf, -np.inf], np.nan, inplace=True)

    # Re-fill NaNs after replacing infinites, just in case
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            df[col].fillna(df[col].mean(), inplace=True)
        else:
            # This case should ideally not be reached if all object columns are encoded
            # but added as a safeguard
            df[col].fillna(df[col].mode()[0], inplace=True)
    print("Infinite values handled and NaNs re-filled.")
else:
    print("No infinite values found. Data remains clean.")

# Final check for NaNs after handling infinites
final_nan_count = df.isnull().sum().sum()
print(f"Final NaN count after handling infinites: {final_nan_count}")

if final_nan_count == 0:
    print("Dataset is now completely free of NaNs and infinites.")
else:
    print("WARNING: NaNs still present. Further inspection required.")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df)

print("Scaling Successful!")
print(scaled_data.shape)
```

```python
import numpy as np

# Check scaled_data for NaNs and infinites
nan_in_scaled_data = np.isnan(scaled_data).sum()
inf_in_scaled_data = np.isinf(scaled_data).sum()

print(f"NaNs found in scaled_data: {nan_in_scaled_data}")
print(f"Infinities found in scaled_data: {inf_in_scaled_data}")

if nan_in_scaled_data > 0 or inf_in_scaled_data > 0:
    print("Handling NaNs/Infinities directly in scaled_data...")
    # Replace infinites with NaNs for easier handling
    scaled_data[np.isinf(scaled_data)] = np.nan

    # Fill NaNs in scaled_data with the column mean
    # Convert to DataFrame temporarily to use fillna efficiently column-wise
    scaled_df = pd.DataFrame(scaled_data)
    scaled_df = scaled_df.fillna(scaled_df.mean())
    scaled_data = scaled_df.to_numpy()
    print("NaNs/Infinities in scaled_data handled by filling with column mean.")
else:
    print("scaled_data is clean (no NaNs or infinites).")

final_nan_check_scaled = np.isnan(scaled_data).sum()
final_inf_check_scaled = np.isinf(scaled_data).sum()
print(f"Final check: NaNs in scaled_data: {final_nan_check_scaled}")
print(f"Final check: Infinities in scaled_data: {final_inf_check_scaled}")
```

```python
sample_df = pd.DataFrame(scaled_data).sample(n=50000, random_state=42)
```

#Elbow method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(1, 11):
    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )
    kmeans.fit(sample_df)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

#K means clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

clusters = kmeans.fit_predict(scaled_data)

df["Cluster"] = clusters

print(df.head())
```

#Cluster Count

 #Cluster Visualization

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)

pca_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(8,6))
plt.scatter(
    pca_data[:,0],
    pca_data[:,1],
    c=clusters,
    cmap="viridis"
)

plt.title("K-Means Clustering")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label="Cluster")
plt.show()
```
